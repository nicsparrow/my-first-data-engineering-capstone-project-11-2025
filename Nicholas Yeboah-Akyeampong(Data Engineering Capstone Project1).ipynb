```python
import random

# Main dictionary to store all accounts
accounts = {}
next_account_number = 100000  #Starting account number

def generate_account_number():
    global next_account_number

    if next_account_number > 999999:
        print("Error: Maximum number of accounts reached!")
        return None

    acc_num = str(next_account_number)
    next_account_number += 1  # move to the next number

    return acc_num


# Create a new account
def create_account():
    print("\n=== Create New Account ===")
    f_name = str(input("Enter your first name: ")).title()
    l_name = str(input("Enter your last name: ")).upper()
    name = f_name + " " + l_name
    initial_deposit = float(input("Enter initial deposit: $"))
        
    account_number = generate_account_number()

    accounts[account_number] = {                        
        "name": name,
        "balance": initial_deposit,
        "transactions": [f"Initial deposit: ${initial_deposit}"]      # Initialize transaction history with the first deposit
    }

    print(f"Account created successfully!")
    print(f"{name}, Welcome to Secure Save Bank!")
    print(f"Account Number: {account_number}")

# Deposit money
def deposit():
    print("\n=== Deposit Money ===")
    acc = input("Enter account number: ")

    if acc not in accounts:
        print("Error: Account not found!")
        return

    amount = float(input("Enter deposit amount: $"))
    accounts[acc]["balance"] += amount
    accounts[acc]["transactions"].append(f"Deposited: ${amount}")

    print("Deposit successful!")
    print(f"New Balance: ${accounts[acc]['balance']}")

# Withdraw money
def withdraw():
    print("\n=== Withdraw Money ===")
    acc = input("Enter account number: ")

    if acc not in accounts:
        print("Error: Account not found!")
        return

    amount = float(input("Enter withdrawal amount: $"))

    if amount > accounts[acc]["balance"]:
        print("Error: Insufficient funds!")
        return

    accounts[acc]["balance"] -= amount
    accounts[acc]["transactions"].append(f"Withdrew: ${amount}")

    print("Withdrawal successful!")
    print(f"New Balance: ${accounts[acc]['balance']}")

# Check balance
def check_balance():
    print("\n=== Check Balance ===")
    acc = input("Enter account number: ")

    if acc not in accounts:
        print("Error: Account not found!")
        return

    print(f"Account Holder: {accounts[acc]['name']}")
    print(f"Current Balance: ${accounts[acc]['balance']}")

# View transaction history
def view_transactions():
    print("\n=== Transaction History ===")
    acc = input("Enter account number: ")

    if acc not in accounts:
        print("Error: Account not found!")
        return

    print(f"Transaction History for Account {acc}:")
    for t in accounts[acc]["transactions"]:
        print(" -", t)

# Menu-driven interface
def menu():
    while True:
        print("\n====== SecureSave Bank System ======")
        print("1. Create Account")
        print("2. Deposit")
        print("3. Withdraw")
        print("4. Check Balance")
        print("5. View Transaction History")
        print("6. Exit")

        choice = input("Select an option: ")

        if choice == "1":
            create_account()
        elif choice == "2":
            deposit()
        elif choice == "3":
            withdraw()
        elif choice == "4":
            check_balance()
        elif choice == "5":
            view_transactions()
        elif choice == "6":
            print("Thank you for using SecureSave Bank!")
            break
        else:
            print("Invalid choice! Please try again.")

# Run the program

menu()
```
